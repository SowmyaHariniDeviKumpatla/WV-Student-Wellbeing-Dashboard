```python
import pandas as pd
import numpy as np
```

```python
# Mock student survey data parameters
years = [2022, 2023, 2024, 2025]
state = "West Virginia"
counties = ["Monongalia", "Kanawha", "Cabell", "Raleigh"]
schools_per_county = {
    "Monongalia": ["Monongalia HS", "University HS"],
    "Kanawha": ["Capital HS", "Riverside HS"],
    "Cabell": ["Cabell Midland HS", "Huntington HS"],
    "Raleigh": ["Woodrow Wilson HS", "Liberty HS"]}
metrics = ["Vaping", "Alcohol Use", "Depression Symptoms", "Bullying", "School Connectedness"]
categories = {
    "Vaping": "Substance Use",
    "Alcohol Use": "Substance Use",
    "Depression Symptoms": "Mental Health",
    "Bullying": "School Climate",
    "School Connectedness": "Protective Factor"}
```

```python
# mock data
data = []
np.random.seed(42) 

for year in years:
    for county in counties:
        for school in schools_per_county[county]:
            for metric in metrics:
                if metric == "School Connectedness":
                    value = np.random.randint(55, 81) 
                else:
                    value = np.random.randint(8, 36) 
                data.append([year, state, county, school, "All Students", metric, categories[metric], value])

df = pd.DataFrame(data, columns=["Year", "State", "County", "School", "Gender", "Metric", "Category", "Value"])

df.to_excel("WVPRC_Mock_Survey_Data.xlsx", index=False)

print("Excel file generated successfully!")
```
